```python
import re
import matplotlib.pyplot as plt

def plot_convergence_curve(log_file):
    """
    绘制训练和验证损失收敛曲线

    参数:
        log_file (str): 日志文件的路径
    """
    # 正则表达式匹配 train_loss 和 val_loss
    train_loss_pattern = re.compile(r"train_loss=(\d+\.\d+)")
    val_loss_pattern = re.compile(r"val_loss=(\d+\.\d+)")

    # 存储 train_loss 和 val_loss
    train_losses = []
    val_losses = []

    # 读取日志文件
    with open(log_file, "r") as file:
        for line in file:
            # 提取 train_loss
            train_match = train_loss_pattern.search(line)
            if train_match:
                train_losses.append(float(train_match.group(1)))
            
            # 提取 val_loss
            val_match = val_loss_pattern.search(line)
            if val_match:
                val_losses.append(float(val_match.group(1)))

    # 绘制收敛曲线
    plt.figure(figsize=(10, 6))
    plt.plot(train_losses, label="Train Loss", color="blue")
    plt.plot(val_losses, label="Validation Loss", color="orange")
    plt.xlabel("Batch Iteration")
    plt.ylabel("Loss")
    plt.title("Train and Validation Loss Convergence")
    plt.legend()
    plt.grid(True)
    plt.show()
```
